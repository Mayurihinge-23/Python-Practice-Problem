```python
n=int(input("Enter the number of employees"))
for i in range(1,n+1):
    salary=int(input(f"Employees {i}:"))
    if salary<30000:
        bonus=salary*0.10
        total=salary+bonus
        print(f"Your salary with bonus {i}:{total}")
    else:
        print(f"Your salary {i}:{salary} ")
        
        
```
