```python
n=5
total=0
average=0

for i in range(1,n+1):
    price=int(input(f"Items {i}: Rs."))
    #total
    total=total+price
    #average
    average=total/n

    if i==1:
        exp=price
        cheap=price
    #expensive
    if price>exp:
        exp=price
    #cheapest    
    if price<cheap:
        cheap=price
print(f"Total price is: Rs.{total}")
print(f"Average of item :{average}")
print(f"Expensive item is {i}:Rs.{exp}")        
print(f"Cheapest item is {i}:Rs.{cheap}") 
```
