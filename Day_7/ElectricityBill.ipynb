```python
#slab calculation
units=int(input("Enter the unit:"))
if units<=100:
    print("Bill= Rs.",units*5)
else:    
    if units >=101 and units<=200:
        print("Bill= Rs.",(100*5)+(units-100)*7)
    else:
        print("Bill= Rs.",(100*5)+(100*7)+(units-200)*10)
```
