```python
n=11
zero=0
positive=0
negative=0
for i in range(1,11):
    num=int(input(f"Enter the num {i}:"))
    if num==0:
        zero+=1
    elif num>=1:
        positive+=1
    else:
        negative+=1
print("Zero are:",zero)        
print("Positive are:",positive)         
print("Negative are:",negative)             
```
