```python
num1=int(input("Enter the number1:"))
operator=input("Operator:")
num2=int(input("Enter the number2:"))
if operator=='+':
    add=num1+num2
    print(f"Addition is: {add}")
elif operator=='-':
    sub=num1-num2
    print(f"Sub is: {sub}")
elif operator=='*':
    mul=num1*num2
    print(f"Multiplication is: {mul}")
elif operator=='/':
    if num2/0:
        print(f"Division by zero is not allowed")
    else:   
        div=num1/num2
        print(f"Division is: {div}")
else:
    print("YOUR CALCULATION IS DONE!!")
    
    
```
