```python
marks=int(input("Enter the marks"))
if(marks>=90 or marks>=100):
    print("Grade A")
elif(marks>=80 or marks>=89):
    print("Grade B")
elif(marks>=70 or marks>=79):
    print("Grade C") 
elif(marks>=60 or marks>=69):
    print("Grade D")   
else:
    print("Grade F")
```
