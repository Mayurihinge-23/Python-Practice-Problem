```python
i=1
secret_num=25
while(i<=1): #1<=1-true
    guess_number=int(input("Guess any number:"))
    if guess_number==secret_num: #23==25 ->false
        print("Congratulations! You guessed correctly.")
        break
    else:    
        print("Wrong Guess!!")
```
