```python
#in While(True) is used for menu should display again and agian after operation complete
i=1
while(True):
    print("1.Addition")
    print("2.Substraction")
    print("3.Multiplication")
    print("4.Division")
    print("5.Exit")
    i=i+1
    perform=int(input("Select any one for operation perform:"))
    if perform==1:
        num1=int(input("Enter num 1:"))
        num2=int(input("Enter num 2:"))
        add=num1+num2
        print(f"Addition is:{add}")
    elif perform==2:
        num1=int(input("Enter num 1:"))
        num2=int(input("Enter num 2:"))
        sub=num1-num2
        print(f"Sub is:{sub}")
    elif perform==3:
        num1=int(input("Enter num 1:"))
        num2=int(input("Enter num 2:"))
        mul=num1*num2
        print(f"Multiplication is:{mul}")
    elif perform==4:
        num1=int(input("Enter num 1:"))
        num2=int(input("Enter num 2:"))
        if num2>0:
            div=num1/num2
            print(f"Division is:{div}")
        else:
            print("Cannot divide by zero")
    elif perform==5:
        break
    else:
        print("In menu You selected is not exist")
   
    
```
