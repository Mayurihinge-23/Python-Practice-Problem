```python
i=1
n=3
PIN=1234
while(i<=n): #1<=3-true
    Pin=int(input("Enter the Pin:"))
    if Pin==PIN: #1236==1234 ->false
        print("Welcome.")
        
        break
    else:    
        print("Wrong PIN!")
        i=i+1
```
