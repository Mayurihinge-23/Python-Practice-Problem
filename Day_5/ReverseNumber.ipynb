```python
n=1234 
rev=0
while n>0:               #1234>0 ->true      |123>0 ->true       |12>0 ->true       |1>0 ->true
    d=n%10               #d=1234%10 ->4      |d=123%10 ->3       |d=12%10 ->2       |d=1%10 ->1
    rev=(rev*10)+d       #rev=(0*10)+4 ->4   |rev=(0*10)+3 ->3   |rev=(0*10)+2 ->2  |rev=(0*10)+1 ->1
    n=n//10              #n=1234//10 ->123   |n=123//10 ->12     |n=12//10 ->1      |n=1//10 ->0  
print("Reverse is:",rev) #print-> 4          |print-> 3          |print-> 2         |print-> 1
```

```python
n=int(input("Enter the number:"))
rev=0
while n>0:
    d=n%10
    rev=(rev*10)+d
    n=n//10
print(f"Reverse is: {rev}")    
```
