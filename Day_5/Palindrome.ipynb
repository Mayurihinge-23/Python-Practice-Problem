```python
n=123
pal=n
rev=0
while n>0:           #123>0 ->true  | #12>0 ->true  | 1>0 ->true
    d=n%10           #123%10 ->3    | #12%10 ->2    | 1%10 ->1
    rev =(rev*10)+d  #rev=3         | #rev=2        | rev=1
    n=n//10          #123//10 ->12  | #12//10 ->1   |  1//10 ->0
if pal==rev: #123==321 ->false
    print("Palindrome")
else:    
    print("Not Palindrome")
```

```python
n=1221
pal=n
rev=0
while n>0:           #1221>0 ->true   | 122>0 ->true   | 12>0 ->true    | 1>0 ->true
    d=n%10           #1221%10 ->1     | 122%10 ->2     | 12%10 ->2      | 1%10 ->1
    rev =(rev*10)+d  #rev=1           | rev=2          | rev=2          | rev=1
    n=n//10          #1221//10 ->122  | 122//10 ->12   | 12//10 ->1     | 1//10 ->0
if pal==rev: #1221 == 1221 ->true
    print("Palindrome")
else:    
    print("Not Palindrome")
```
