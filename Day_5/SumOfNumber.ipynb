```python
n=1234
sum=0
while n>0: #1234>0 ->true
    d=n%10 #d=1234%10 ->4
    sum+=d  #sum=4
    n=n//10 #1234//10 ->123
print(f"Sum is:{sum}")    
    
    
```

```python
n=int(input("Enter the number:"))
sum=0
while n>0: #123>0 ->true
    d=n%10 #123%10 ->3
    sum+=d #sum=4
    n=n//10 #123//10 ->12
print(f"Reverse is: {sum}")    
```

```python
n=int(input("Enter the number:"))
sum=0
while n>0: #1456>0 ->true      |145>0 ->true    |14>0 ->true     |1>0 ->true
    d=n%10 #1456%10 ->6        |145%10 ->5      |14%10 ->4       |1%10 ->1
    sum+=d #sum=6              |sum=6+5 =11     |sum=6+5+4 =15   |sum=6+5+4+1 =16
    n=n//10 #1456//10 ->145    |145//10 ->14   |14//10 ->1       |1//10 ->0
print(f"Reverse is: {sum}")    
```
