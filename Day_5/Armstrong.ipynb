```python
n=153
arm=n
sum=0
while n>0:             #153>0 ->true       #15>0               #1>0
    d=n%10             #153%10 ->3         #15%10 ->5          #1%10 ->1
    c=d**3             #3**3=27            #5**3 ->125         #1**3=1
    sum=sum+c          #0+27=27            #27+125 =152        #152+1=153
    n=n//10            #153//10 ->15       #15//10 ->1         #1//10 ->0
if arm==sum:           #153==27 ->false    #153==152->false    #153= 153 ->true
    print("Armstrong")
else:
    print("Not Armstrong")
```
