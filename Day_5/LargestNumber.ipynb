```python
n=231
lar=0
while n>0:       #231>0 ->true    #23>0 ->true    --2>0->true
    d=n%10       #231%10=1        #23%10 ->3      --2%10->2
    if d>lar:    #1>0 ->true      #3>1->true      --2>3 ->false
        lar=d    #lar=1           #lar=3          --lar=3
    n=n//10      #231//10 ->23    #23//10 ->2     --2//10 ->0
print(f"Largest is:{lar}")    
```
