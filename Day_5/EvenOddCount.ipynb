```python
n=1233
even=0
odd=0
while n>0:       #1233>0 ->true    -- 123>0 ->true   -- 12>0 ->true   -- 1>0 ->true
    d=n%10           #1233%10 ->3      -- 123%10 ->3     -- 12%10 ->2     -- 1%10 ->1
    if d%2==0:       #3%2==0 ->false   --3%2==0 ->false  --2%2==0 ->true  --1%2==0 ->false
        even=even+1                                      #even=0+1=1
    else:            #3%2!=0 ->true    --3%2!=0 ->true                    --1%2!=0 ->true
        odd=odd+1    #odd=0+1 ->1      --odd=1+1 ->2                      --odd=2+1 ->3
    n=n//10          #1233//10 ->123   --123//10 ->12    --12//10 ->1     --1//10 ->0
print(f"Even are:{even}")    
print(f"Odd are:{odd}")           
```
