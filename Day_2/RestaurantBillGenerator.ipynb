```python
original_bill=float(input("Enter the bill amount:"))
# discount=int(input("Discount is:"))
# final_bill=int(input("Final bill is:"))

discount=0
final_bill=0

if original_bill >=5000:#1500>5000 -> false
    discount=original_bill*0.20
    final_bill=original_bill-discount
    print(f"Your Final bill amount is: {final_bill}")
elif original_bill >=3000 or original_bill >=4999:#1500>3000 ->false or 1500>4999->false
    discount=original_bill*0.10
    final_bill=original_bill-discount
    print(f"Your Final bill amount is: {final_bill}")
elif original_bill <3000:#1500<3000
    print(f"Sorry below 3000 no discount /n Your Final bill is: {original_bill}")
else:
    print("Thank you for shopping")
```
