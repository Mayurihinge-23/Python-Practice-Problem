```python
age=int(input("Enter Your age:"))
# if age<18:
#     print(f"Sorry you cannot book hotel \n Reason:Your age is not valid")
    
has_Id=input("Is you have ID  :Yes/No")
if age>=18 and has_Id=='Yes':
    print("Booking Confirmed")
elif age>=18 and has_Id=='No':
    print(f"Booking Rejected \nValid ID is required")
elif age<18 and has_Id=='Yes':
    print("Booking Rejected \n Reason:Customer must be at least 18 years old")
elif age<18 and has_Id=='No':
    print("Booking Rejected \n Reason:Customer must be at least 18 years old and have Valid Id")    
else:
    print("ThankYou for visiting")
```
