```python
Basic_salary=int(input("Enter the Salary:"))
HRA=Basic_salary*0.20 #HRA=20% on basic_salary
DA=Basic_salary*0.10  #DA=10% on basic_salary
Gross_salary=Basic_salary+HRA+DA  #GS=40000+8000+4000=52000

if Basic_salary>50000: #40000>50000 -> false
    Tax_deduct=2000 
    Net_salary=Gross_salary-Tax_deduct
    print(f"Net salary of employee:{Net_salary}")
else: #40000<50000 -> true
    Tax_deduct=1000 
    Net_salary=Gross_salary-Tax_deduct #52000-1000=51000
    print(f"Net salary of employee:{Net_salary}")
```
