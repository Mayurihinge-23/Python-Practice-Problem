```python
Luggage_weight=int(input("Enter the weight of Luggage in Kg:"))
if Luggage_weight<=0 or Luggage_weight<=15: #0<=0 ->true or 0<=15->true
    print("No extra charges required")
elif Luggage_weight<=16 or Luggage_weight<=25:
    print("Extra charges required Rs.500")  
else:
    print("Extra charges required  Rs.1000")
```
