```python
car_name=input("Enter your Car Number:")

hours=int(input(f"Car parking hours?:"))
print("You have to pay charges according to parking hours!")
if hours<=1 or hours<=2:
    charges=hours*30
    print(f"Your parking charges is: {charges} for {hours}hrs")
    
elif hours<=3 or hours<=4:
    charges=hours*20
    print(f"Your parking charges is: {charges} for {hours}hrs")
    
else:        
    charges=hours*15
    print(f"Your parking charges is: {charges} for {hours}hrs")
```
