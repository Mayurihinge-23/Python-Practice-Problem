```python
age=int(input("Enter the age:"))
if age<5: #4<5 -> True
    print("No Ticket")
elif age>=5 and age<=17:
    print("Ticket is Rs.100")
elif age>=18 and age<=59:
    print("Ticket is Rs.250")
else:
    print("Ticket is Rs.150")
```
