```python
name=input("Enter your name:")

last_day=int(input(f"You are submitting after days:"))
print("You have to pay fine Rs.2/day")
if last_day<=1 or last_day<=5:
    fine=last_day*2
    print(f"You have to pay fine after {last_day} days is: Rs.{fine}")
    
elif last_day<=6 or last_day<=10:
    fine=last_day*5
    print(f"You have to pay fine after {last_day} days is: Rs.{fine}")
    
else:        
    fine=last_day*10
    print(f"You have to pay fine after {last_day} days is: Rs.{fine}")
```
