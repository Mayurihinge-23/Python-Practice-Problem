```python
marks=int(input("Enter the marks of student:"))
practical_marks=int(input("Enter the practical marks of student:"))
if marks>=40 and practical_marks>=20: #35>=40 -> false and 15>=20 ->false  (Failed)
    print("Result:Passed")
    if marks>75:
        print("Remark: Distinction")
    elif marks>=65:
        print("Remark: First Class")
    elif marks>=50:
        print("Remark: Second class")
    else:
        print("Remark: Pass")
else:
    print("Failed")
```
