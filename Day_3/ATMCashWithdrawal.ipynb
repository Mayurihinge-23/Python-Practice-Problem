```python
PIN=int(input("Enter the PIN:"))
# Account_bal=int(input("Enter the Account balance:"))   
# Withdraw_amt=int(input("Enter the Withdrawal Amount:"))  
if PIN==1234:
    print("PIN is correct")
    Account_bal=int(input("Enter the Account balance:"))   
    Withdraw_amt=int(input("Enter the Withdrawal Amount:"))
    if Withdraw_amt>=1: 
        if Account_bal>=Withdraw_amt:
            Available_bal=Account_bal-Withdraw_amt
            print(f"You have available balance : {Available_bal}")
        elif Account_bal<Withdraw_amt:  
            print("Sorry!! You have Insufficient balance")
        else:
            print("Please!! Check your balance")
    else:
        print("Sorry you cannot enter 0 withdraw amount")
else:
    print("Sorry You have enter incorrect PIN") 
```
