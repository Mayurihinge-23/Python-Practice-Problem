```python
salary=int(input("Enter the salary:"))
experience=int(input("How many year of experience you have:"))
if salary>=50000 and experience>=5:
    bonus=salary*0.20
    total_bonus=salary+bonus
    print(f"You have bonus 20%. \nTotal salary is: {total_bonus}")
elif salary>=30000 and experience>=3:
    bonus=salary*0.10
    total_bonus=salary+bonus
    print(f"You have bonus 10%. \nTotal salary is: {total_bonus}")
else:
    print("No bonus")
```
