```python
Class=input("Your class please:")
Luggage_weight=int(input("Enter the luggage weight in Kg:"))
if Class=='Economy':
    if Luggage_weight<=15: #15<=15->true
        print("You have no charges")
    elif Luggage_weight<=35:
        print("You will required Rs.500 charges above 15Kg")
    else:
        print("Boarding Denied")
else:
    if Class=='Bussiness':
        if Luggage_weight<=25: #15<=15->true
            print("You have no charges")
        elif Luggage_weight<=35:
            print("You will required Rs.500 charges above 25Kg")
        else:
            print("Boarding Denied")
```
