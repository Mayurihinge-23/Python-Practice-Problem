```python
Percentage=int(input(f"Your 12th percentage:"))
CET=int(input(f"Your CET marks:"))
if Percentage>=60 and CET>=75: #61>=60 ->true and 77>=75->true
    if Percentage>=90: #61>=90 -> false
        print("You are eligible for Computer Science")
    elif Percentage>=80:  #61>=80 -> false
        print("You are eligible for IT")
    elif Percentage>=70:  #61>=70 -> false
        print("You are eligible for Electronics")
    else:
        print("Sorry! Our criteria isn't matching to your marks")
else:
    print("Addmission Rejected")        
```
