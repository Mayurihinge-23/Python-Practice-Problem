```python
age=int(input("Enter your age"))
# salary=int(input("Enter your salary"))
# credit_score=int(input("Enter your credit score"))
if age>=21  and age<=60: #21>=21 ->true and 21<=60 ->true
    print("Yes! Age is Valid for loan")
    salary=int(input("Enter your salary"))  
    credit_score=int(input("Enter your credit score"))
    if salary>=30000 and credit_score>=700: #30000>=30000 ->true and 700>=700->true
        print(f"Your credit score is: {credit_score} \nLoan Approved") 
    elif credit_score<700:
        print("Credit score is low")
    else:
        print("Salary is low")
else:
    print("Age is Invalid")
            
```
