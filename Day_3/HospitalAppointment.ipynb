```python
age=int(input("Enter the age:"))
Fever=input("Is you have fever (Yes/No):")
chest_pain=input("Is you have chest pain (Yes/No):")
if age<18:
    print("Assigned to child doctor")
elif age>18 and Fever=='Yes' or chest_pain=='Yes':
    print("Assigned to General doctor")
# elif Fever=='Yes' or chest_pain=='Yes' :
    print("Emergency Consultation Required")
else:
    print("Don't Worry! Its Normal")
    
```
